## 3. ***Data Wrangling***

````python
# =========================================
# DATA WRANGLING
# =========================================

import pandas as pd
import numpy as np

# =========================================
# LOAD DATASET
# =========================================

df = pd.read_csv('data_YesBank_StockPrices.csv')

# Display first 5 rows
print(df.head())

# =========================================
# DATASET SHAPE
# =========================================

print("\nDataset Shape:", df.shape)

# =========================================
# DATASET INFORMATION
# =========================================

print("\nDataset Info:\n")
print(df.info())

# =========================================
# CHECK MISSING VALUES
# =========================================

print("\nMissing Values:\n")
print(df.isnull().sum())

# =========================================
# HANDLE MISSING VALUES
# =========================================

# Fill numerical missing values with median

for column in df.select_dtypes(include=np.number).columns:
    df[column].fillna(df[column].median(), inplace=True)

print("\nMissing Values After Treatment:\n")
print(df.isnull().sum())

# =========================================
# CHECK DUPLICATES
# =========================================

print("\nDuplicate Values:", df.duplicated().sum())

# Remove duplicates
df.drop_duplicates(inplace=True)

print("\nDuplicates Removed Successfully")

# =========================================
# CONVERT DATE COLUMN
# =========================================

# Convert Date column into datetime format

df['Date'] = pd.to_datetime(df['Date'], format='%b-%y')

print("\nDate Column Converted Successfully")

# =========================================
# CHECK DATA TYPES
# =========================================

print("\nData Types:\n")
print(df.dtypes)

# =========================================
# SORT DATASET BY DATE
# =========================================

df = df.sort_values(by='Date')

print("\nDataset Sorted by Date")

# =========================================
# RESET INDEX
# =========================================

df.reset_index(drop=True, inplace=True)

print("\nIndex Reset Successfully")

# =========================================
# FINAL DATASET SHAPE
# =========================================

print("\nFinal Dataset Shape:", df.shape)

# =========================================
# DISPLAY CLEANED DATASET
# =========================================

print("\nCleaned Dataset:\n")

df.head()
````

## ***5. Hypothesis Testing***

**Statement 1**

    “There is a significant relationship between Open Price and Close Price.”

Statistical Test Used

    Pearson Correlation Test

Hypotheses

    Null Hypothesis (H₀): No significant relationship exists between Open and Close prices.

    Alternative Hypothesis (H₁): A significant relationship exists between Open and Close prices.

````python
from scipy import stats
import numpy as np
````

````python
# Pearson Correlation Test

correlation, p_value = stats.pearsonr(df['Open'], df['Close'])

print("Correlation Coefficient:", correlation)
print("P-value:", p_value)
````

````python
#Conclusion Rule
if p_value < 0.05:
    print("Reject Null Hypothesis")
    print("There is a significant relationship between Open and Close Price.")
else:
    print("Fail to Reject Null Hypothesis")
````

**Statement 2**

    “The average High Price is significantly greater than the average Low Price.”

Statistical Test Used

    Paired T-Test

Hypotheses

    Null Hypothesis (H₀): Mean High Price equals Mean Low Price.

    Alternative Hypothesis (H₁): Mean High Price is significantly greater than Mean Low Price.

````python
# Paired T-Test

t_stat, p_value = stats.ttest_rel(df['High'], df['Low'])

print("T-Statistic:", t_stat)
print("P-value:", p_value)
````

````python
#Conclusion Rule
if p_value < 0.05:
    print("Reject Null Hypothesis")
    print("Average High Price is significantly greater than Average Low Price.")
else:
    print("Fail to Reject Null Hypothesis")
````

Statement 3

    “There is significant volatility in Yes Bank stock prices over time.”

Statistical Test Used

    Variance Analysis

Hypotheses

    Null Hypothesis (H₀): Stock prices show no significant volatility.

    Alternative Hypothesis (H₁): Stock prices show significant volatility.



````python
# Calculate variance of Closing Price

variance = np.var(df['Close'])

print("Variance of Closing Price:", variance)
````

````python
#Additional Volatility Measure
# Standard Deviation

std_dev = np.std(df['Close'])

print("Standard Deviation:", std_dev)
````

Final Interpretation

    -Low p-value (< 0.05) indicates statistically significant results.

    -Strong correlation between Open and Close prices supports predictive modeling.

    -Significant difference between High and Low prices confirms stock price fluctuations.

    -High variance and standard deviation indicate market volatility in Yes Bank stock prices.

##Handling Missing Values

````python
#Handling Missing Values
# Check missing values

df.isnull().sum()
````

````python
# Fill missing numerical values using median

for column in df.select_dtypes(include=np.number).columns:
    df[column].fillna(df[column].median(), inplace=True)

# Verify missing values again
df.isnull().sum()
````

##Handling Outliers

````python
#Handling Outliers
# Boxplot for detecting outliers

plt.figure(figsize=(10,6))

sns.boxplot(data=df[['Open', 'High', 'Low', 'Close']])

plt.title("Outlier Detection")
plt.show()
````

````python
# IQR Method

Q1 = df[['Open','High','Low','Close']].quantile(0.25)
Q3 = df[['Open','High','Low','Close']].quantile(0.75)

IQR = Q3 - Q1

print(IQR)
````

````python
# Remove Outliers

df = df[~((df[['Open','High','Low','Close']] < (Q1 - 1.5 * IQR)) |
          (df[['Open','High','Low','Close']] > (Q3 + 1.5 * IQR))).any(axis=1)]

print(df.shape)
````

## ***7. ML Model Implementation***

````python
#Import Required Libraries
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from sklearn.model_selection import GridSearchCV

import matplotlib.pyplot as plt
import numpy as np
````

## MODEL - 1 : Linear Regression

````python
#Fit the Algorithm
# Create model

lr_model = LinearRegression()

# Train model

lr_model.fit(X_train, y_train)
````

````python
#Predict on the Model
# Prediction

lr_pred = lr_model.predict(X_test)
````

````python
#Evaluation Metrics
# Evaluation

lr_mae = mean_absolute_error(y_test, lr_pred)
lr_rmse = np.sqrt(mean_squared_error(y_test, lr_pred))
lr_r2 = r2_score(y_test, lr_pred)

print("MAE :", lr_mae)
print("RMSE :", lr_rmse)
print("R2 Score :", lr_r2)
````

````python
#Visualizing Evaluation Metric Score Chart
metrics = ['MAE', 'RMSE', 'R2 Score']
scores = [lr_mae, lr_rmse, lr_r2]

plt.figure(figsize=(8,5))

plt.bar(metrics, scores)

plt.title("Linear Regression Performance")

plt.show()
````

Linear Regression is a supervised Machine Learning algorithm used for predicting continuous numerical values. It establishes a linear relationship between independent variables and the target variable. In this project, Open, High, and Low prices are used to predict the Close price of Yes Bank stock.

The model performed well because stock price variables are highly correlated, making Linear Regression suitable for financial prediction tasks.

````python
#Cross Validation & Hyperparameter Tuning
# Parameters

params = {
    'fit_intercept': [True, False]
}

# GridSearchCV

grid_lr = GridSearchCV(
    LinearRegression(),
    params,
    cv=5,
    scoring='r2'
)

# Fit model

grid_lr.fit(X_train, y_train)
````

````python
#Best Parameters
print(grid_lr.best_params_)
````

````python
#Prediction after Tuning
lr_best_pred = grid_lr.predict(X_test)

print("Tuned R2 Score:",
      r2_score(y_test, lr_best_pred))
````

GridSearchCV was used for hyperparameter tuning because it systematically searches all possible parameter combinations and selects the best-performing model based on cross-validation performance. It helps improve model accuracy and reliability.

Yes, slight improvement was observed in the R² score after hyperparameter tuning because the optimized parameters improved the model fitting process.
